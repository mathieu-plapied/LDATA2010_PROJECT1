```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import plotly.express as px

PATH = "data/transcriptomics_data.csv"
```

```python
df = pd.read_csv(PATH)
```

```python
features = list(df.columns)
features.remove("colour")
features.remove("cell_type")
```

```python
x = df.loc[:, features].values
y = df.loc[:, ['cell_type']].values

#x = StandardScaler().fit_transform(x)
```

```python
pca = PCA(n_components=2)
principalComponnents = pca.fit_transform(x)
principalDf = pd.DataFrame(data=principalComponnents, columns=["pca1", "pca2"])
finalDf = pd.concat([principalDf, df[["cell_type", "colour"]]], axis=1)
```
